```python
file_path = '/content/drive/My Drive/dataset/Social_Network_Ads.csv'
```

```python
data = pd.read_csv(file_path)
```

```python
data.head()
```

```python
data = data.drop(columns="User ID")
```

```python
data = pd.get_dummies(data)
```

```python
data.head()
```

```python
x = data.drop(columns="Purchased")
y = data.Purchased
```

```python
xi = data.drop(columns="Purchased")
```

```python
#STANDARISASI
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(x)
data_scaled = scaler.transform(x)
data_scaled = pd.DataFrame(data_scaled, columns=x.columns)
data_scaled.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
minscaler = MinMaxScaler()
mmscaler = minscaler.fit_transform(xi)
mmscaler = pd.DataFrame(mmscaler, columns=xi.columns)
mmscaler.head()
```

```python
#pakai yang standard scaling
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(data_scaled, y, test_size=0.2, random_state=42)
```

```python
#definisikan model
from sklearn import linear_model
modellog = linear_model.LogisticRegression()
modellog.fit(x_train, y_train)
```

```python
#uji data train
modellog.score(x_test, y_test)
```
